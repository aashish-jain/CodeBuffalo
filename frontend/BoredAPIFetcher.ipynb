```python
from requests import get
import json
```

```python
dic = {}
```

```python
for i in range(10000):
    resp = get("http://www.boredapi.com/api/activity/")
    row = json.loads(resp.content)
    dic[row['key']] = row
```

```python
len(dic)
```

```python
dic
```

```python
json.dump(dic, open("BoredAPIDump.json","w"))
```
